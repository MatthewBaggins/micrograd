```python
import math
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

from micrograd.value import Value
from micrograd.draw import draw_dot
```

```python
x1 = Value(2, "x1")
x2 = Value(0, "x2")

w1 = Value(-3, "w1")
w2 = Value(1, "w2")

b = Value(6.88, "b")

x1w1 = x1*w1 | "x1 * w1"
x2w2 = x2*w2 | "x2 * w2"

x1w1x2w2 = x1w1 + x2w2 | "x1w1 + x2w2"

n = x1w1x2w2 + b | "n"

e = (2*n).exp()

# o = n.tanh() | "o"
o = (e - 1) / (e + 1) | "o"
o.backprop()
draw_dot(o)
```
